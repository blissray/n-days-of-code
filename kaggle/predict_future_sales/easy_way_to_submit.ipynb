```python
import pandas as pd
```

```python
item_category_df = pd.read_csv("data/item_categories.csv")
items_df = pd.read_csv("data/items.csv")
shops_df = pd.read_csv("data/shops.csv")

train_df = pd.read_csv("data/sales_train.csv")
test_df = pd.read_csv("data/test.csv")
```

```python
train_df.head()
```

```python
date_df = pd.DataFrame(train_df["date"].str.split(".").tolist(),  columns = ['day','month','year'])
date_df.head()
```

```python
pre_df = pd.merge(train_df, date_df, left_index=True, right_index=True)
del pre_df["date"]
del pre_df["date_block_num"]
pre_df.head()
```

```python
pre_df.groupby(["shop_id","item_id","item_price"]).mean()
```

# The easiest way to submit

```python
test_df.head()
```

```python
(test_df["shop_id"].map(lambda x : str(x)) + "-" + test_df["item_id"].map(lambda x : str(x))).head()
```

```python
easy_result = pre_df.groupby(["shop_id","item_id"]).mean().reset_index()
easy_result.head()
```

```python
easy_result["new_id"] = easy_result["shop_id"].map(lambda x : str(x)) + "-" + easy_result["item_id"].map(lambda x : str(x))
```

```python
del easy_result ["shop_id"]
del easy_result ["item_id"]
del easy_result ["item_price"]

```

```python
easy_result["new_id"]
```

```python
easy_test = pd.DataFrame(
    test_df["shop_id"].map(lambda x : str(x)) + "-" + test_df["item_id"].map(lambda x : str(x)), columns=["new_id"])
```

```python
easy_test.head()
```

```python
len(easy_test)
```

```python
len(easy_result)
```

```python
final_result =pd.merge(easy_test, easy_result, how="left", left_on="new_id", right_on="new_id")
final_result.head()
```

```python
del final_result["new_id"]
```

```python
final_result[final_result.isnull()]  = 0
final_result.isnull().sum()
```

```python
final_result.index.name = 'ID'
final_result.rename(index=str, columns={"item_cnt_day": "item_cnt_month"}, inplace=True)
final_result.head()
```

```python
final_result.to_csv("easy_result.csv", sep=",")
```
